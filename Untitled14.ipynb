**jelaskan struktur dataset tersebut (jumlah baris, kolom, tipe data)!**



*   **Jumlah baris** dataset Top 100 Worlds Largest Cities memiliki **100 baris data**. Dengan menggunakan perintah **data.shape**, menunjukkan bahwa terdapat 100 catatan yang tersimpan dalam dataset tersebut. Setiap baris mewakili satu kota terbesar di dunia beserta informasi yang berkaitan dengannya, jadi dataset ini memuat data lengkap untuk 100 kota.
*  **Kolom** dataset Top 100 Worlds Largest Cities memiliki **5 kolom** yang masing-masing berisi informasi berbeda tentang setiap kota. Setiap kolom mewakili satu jenis data atau variabel yang menjelaskan karakteristik kota tersebut.

    Kelima kolom tersebut adalah Rank , City , Country , Population (Est.) , dan Area (sq km) . Kolom Rank berisi nomor peringkat kota. Kolom Kota berisi nama kota, sedangkan kolom Negara berisi nama negara tempat kota tersebut berada. Kolom Population (Est.) menunjukkan jumlah perkiraan penduduk di kota tersebut, dan kolom Area (sq km) menunjukkan luas wilayah kota dalam satuan kilometer persegi.
    Dengan adanya lima kolom ini, setiap baris dalam dataset memiliki informasi yang lengkap mengenai identitas dan ukuran masing-masing kota.

*   **Tipe data** dataset Top 100 Worlds Largest Cities memiliki dua jenis tipe data, yaitu int64 dan object.dengan menggunakan perintah **data.info().** Kolom Rank bertipe **int64** , yang berarti berisi angka bulat karena digunakan sebagai nomor peringkat kota.

    Sementara itu, kolom City , Country , Population (Est.) , dan Area (sq km) bertipe **objek** . Tipe ini biasanya digunakan untuk data teks. Kolom City dan Country memang berupa teks, sedangkan Population (Est.) dan Area (sq km) terbaca sebagai teks karena mengandung tanda koma pada angka.

    Secara keseluruhan, dataset ini memiliki satu kolom numerik dan empat kolom bertipe objek. Untuk melakukan perhitungan, kolom populasi dan luas wilayah perlu diubah terlebih dahulu ke tipe numerik.






```python
import json

# Membuat data JSON secara manual
data_json = [
    {
        "Rank": 1,
        "City": "Tokyo",
        "Country": "Japan",
        "Population": 37115000
  },
  {
    "Rank": 2,
    "City": "Delhi",
    "Country": "India",
    "Population": 35500000
  },
  {
    "Rank": 3,
    "City": "Shanghai",
    "Country": "China",
    "Population": 31050000
  }
]

# Menampilkan dalam format JSON yang rapi
print(json.dumps(data_json, indent=4))
```

```python
import pandas as pd

# Ubah JSON menjadi DataFrame
data = pd.DataFrame(data_json)

# Simpan menjadi file CSV
data.to_csv("data_kota.csv", index=False)

# Tampilkan hasil
data
```
